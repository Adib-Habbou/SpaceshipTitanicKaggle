## Importation des donn√©es

```python
import pandas as pd
import numpy as np
train = pd.read_csv("train.csv")
```

```python
train.head(10)
```

```python
train.Cabin.fillna("X/9999/X", inplace = True)
```

```python
train.insert(4, 'CabinDeck', train.Cabin.apply(lambda x: str(x).split("/")[0]))
train.insert(5, 'CabinNum', train.Cabin.apply(lambda x: str(x).split("/")[1]))
train.insert(6, 'CabinSide', train.Cabin.apply(lambda x: str(x).split("/")[2]))
```

```python
train.drop(["Cabin"], axis = 1, inplace = True)
```

```python
train.head(10)
```

## Valeurs Manquantes

### Home Planet

```python
train.HomePlanet.isna().sum()
```

```python
train.groupby("HomePlanet").HomePlanet.count()
```

```python
for index in train.index:
  row = train.iloc[index]
  if (pd.isnull(row.HomePlanet)):
    if (row.CabinDeck in ["A", "B", "C", "T"]):
      train.loc[index, "HomePlanet"] = "Europa"
    if (row.CabinDeck in ["D", "E"]):
      train.loc[index, "HomePlanet"] = "Mars"
    if (row.CabinDeck in ["F", "G", "X"]):
      train.loc[index, "HomePlanet"] = "Earth"
```

```python
train.HomePlanet.isna().sum()
```

### CryoSleep

```python
train.CryoSleep.isna().sum()
```

```python
for index in train.index:
  row = train.iloc[index]
  if (pd.isnull(row.CryoSleep)):
    if (row.HomePlanet == "Mars" or row.RoomService != 0 or row.FoodCourt != 0 or row.ShoppingMall != 0 or row.Spa != 0 or row.VRDeck != 0):
      train.loc[index, "CryoSleep"] = False
    else: train.loc[index, "CryoSleep"] = True
```

### VIP

### RoomService, FoodCourt, ShoppingMall, Spa, VRDeck

```python
sum([train.RoomService.isna().sum(), train.FoodCourt.isna().sum(), train.ShoppingMall.isna().sum(), train.Spa.isna().sum(), train.VRDeck.isna().sum()])
```

```python
meanRoomService = train.RoomService[train.CryoSleep == False].mean()
meanFoodCourt = train.FoodCourt[train.CryoSleep == False].mean()
meanShoppingMall = train.ShoppingMall[train.CryoSleep == False].mean()
meanSpa = train.Spa[train.CryoSleep == False].mean()
meanVRDeck = train.VRDeck[train.CryoSleep == False].mean()
```

```python
for index in train.index:
  row = train.iloc[index]
  if (pd.isnull(row.RoomService)):
    if (row.CryoSleep == True or row.Age <= 12): train.loc[index, "RoomService"] = 0
    else: train.loc[index, "RoomService"] = meanRoomService
  if (pd.isnull(row.FoodCourt)):
    if (row.CryoSleep == True or row.Age <= 12): train.loc[index, "FoodCourt"] = 0
    else: train.loc[index, "FoodCourt"] = meanFoodCourt
  if (pd.isnull(row.ShoppingMall)):
    if (row.CryoSleep == True or row.Age <= 12): train.loc[index, "ShoppingMall"] = 0
    else: train.loc[index, "ShoppingMall"] = meanShoppingMall
  if (pd.isnull(row.Spa)):
    if (row.CryoSleep == True or row.Age <= 12): train.loc[index, "Spa"] = 0
    else: train.loc[index, "Spa"] = meanSpa
  if (pd.isnull(row.VRDeck)):
    if (row.CryoSleep == True or row.Age <= 12): train.loc[index, "VRDeck"] = 0
    else: train.loc[index, "VRDeck"] = meanVRDeck
```

```python
sum([train.RoomService.isna().sum(), train.FoodCourt.isna().sum(), train.ShoppingMall.isna().sum(), train.Spa.isna().sum(), train.VRDeck.isna().sum()])
```

### Cabin Deck



```python
import random
for index in train.index:
  row = train.iloc[index]
  if (row.CabinDeck == "X"):
    if (row.HomePlanet == "Earth"):
      train.loc[index, "CabinDeck"] = random.choice(["F", "G"])
    if (row.HomePlanet == "Mars"):
      train.loc[index, "CabinDeck"] = random.choice(["D", "E"])
    if (row.HomePlanet == "Europa"):
      train.loc[index, "CabinDeck"] = random.choice(["A", "B", "C", "T"])
```

### Cabin Side

```python
import random
for index in train.index:
  row = train.iloc[index]
  if (row.CabinSide == "X"):
      train.loc[index, "CabinSide"] = random.choice(["P", "S"])
```

## Feature Engineering

### Expenses

```python
train.insert(train.columns.get_loc("Name"), "Expenses", train[["RoomService", "FoodCourt", "ShoppingMall", "Spa", "VRDeck"]].sum(axis=1))
```

```python
train.head()
```

### Spending

```python
train.insert(train.columns.get_loc("Name"), "Spending", train["Expenses"].apply(lambda x: 0 if x == 0 else 1))
```

```python
train.head()
```

### AgeRange

```python
def age_range(age):
    if age <= 12:
        return 0
    elif age <= 18:
        return 1
    elif age <= 25:
        return 2
    else:
        return 3
```

```python
train.insert(train.columns.get_loc("VIP"), "AgeRange", train["Age"].apply(age_range))
```

```python
train.head()
```

### Solo

```python
def is_solo(passenger_id):
    group = passenger_id.split("_")[0]
    if (len(train[train["PassengerId"].str.startswith(group)]) == 1):
      return 1
    return 0
```

```python
train.insert(train.columns.get_loc("HomePlanet"), "Solo", train["PassengerId"].apply(is_solo))
```

```python
train.head()
```

### FamilySize

```python
def get_family_size(name):
    last_name = name.split(" ")[1]
    last_name_counts = train["Name"].apply(lambda x: x.split(" ")[1]).value_counts().to_dict()
    family_size = last_name_counts.get(last_name, 1)
    return family_size
```

```python
train.insert(train.columns.get_loc("HomePlanet"), "FamilySize", train["Name"].apply(get_family_size))
```
